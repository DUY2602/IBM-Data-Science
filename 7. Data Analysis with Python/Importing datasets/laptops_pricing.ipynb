```python
import pandas as pd
import numpy as np
import requests
from io import StringIO
```

```python
def download(url, filename):
    response = requests.get(url)
    if response.status_code == 200:
        with open ('laptops.csv', 'w', encoding='utf-8') as f:
            f.write(response.text)
    else:
        print('Download failed: {response.status_code}')

url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-Coursera/laptop_pricing_dataset_base.csv"
filename = "laptops.csv"

download(url, filename)

df = pd.read_csv(filename, header=None)
```

```python
# 1. Print the first 5 rows to verify right data
df.head()
```

```python
# 2. Add headers to the Dataframe
headers = [ "Manufacturer", "Category", "Screen", "GPU", "OS", "CPU_core", "Screen_Size_inch", "CPU_frequency", "RAM_GB", "Storage_GB_SSD", "Weight_kg", "Price"]

df.columns = headers

df.head(10)
```

```python
# 3. Replace ? with NaN and delete those records
df.replace("?", np.NaN)
df.dropna()
```
