```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://web.archive.org/web/20230224123642/https://www.ibm.com/us-en/"

data = requests.get(url).text

soup = BeautifulSoup(data, "html.parser")

for link in soup.find_all("a", href = True):
    print(link.get("href"))
```

```python
for link in soup.find_all("img"):
    print(link.get("src"))
```

```python
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/datasets/HTMLColorCodes.html"

data = requests.get(url).text

soup = BeautifulSoup(data, "html.parser")

table = soup.find("table")
```

```python
url = "https://en.wikipedia.org/wiki/World_population"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 11.0; Win64; x64) "
                  "AppleWebKit/537.36 (KHTML, like Gecko) "
                  "Chrome/91.0.4472.124 Safari/537.36"
}

data  = requests.get(url, headers=headers)

soup = BeautifulSoup(data.text, "html.parser")

tables = soup.find_all("table")
len(tables)
```

```python
for index,table in enumerate(tables):
    if ("10 most densely populated countries" in str(table)):
        table_index = index
print(table_index)
```

```python
population_data = pd.DataFrame(columns=["Rank", "Country", "Population", "Area", "Density"])

for row in tables[table_index].find_all("tr"):
    cols = row.find_all("td")

    if cols:
        rank = cols[0].text.strip()
        country = cols[1].text.strip()
        population = cols[2].text.strip()
        area = cols[3].text.strip()
        density = cols[4].text.strip()

        new_row = pd.DataFrame([{"Rank": rank, "Country": country, "Population": population, "Area": area, "Density": density}])

        population_data = pd.concat([population_data, new_row], ignore_index=True)

population_data
```
